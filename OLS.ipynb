```python
import statsmodels.api as sm
import pandas as pd
```

```python
news_sentiment = pd.read_csv('assets/news_sentiment.csv')
news_sentiment
```

```python
stock_prices = pd.read_csv('assets/stock_prices.csv')
stock_prices
```

```python
apple_stock_prices = stock_prices[stock_prices['stock'] == 'Apple']
apple_stock_prices
```

```python
apple_news_sentiment = news_sentiment[news_sentiment['stock'] == 'Apple']
apple_news_sentiment
```

```python
merged = pd.merge(apple_stock_prices, apple_news_sentiment, on = ['stock', 'date'], how = 'inner')
```

```python
merged
```

```python
y = merged['close'] # dependent variable
X = merged[['avg_sentiment', 'volume']] # independent variable
```

```python
X = sm.add_constant(X)
```

```python
model = sm.OLS(y, X)
result = model.fit()
```

```python
print(result.summary())
```
