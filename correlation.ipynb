```python
import pandas as pd
```

```python
stock_prices = pd.read_csv('assets/stock_prices.csv')
stock_prices
```

```python
news_sentiment = pd.read_csv('assets/news_sentiment.csv')
news_sentiment
```

```python
news_sentiment['date'] = pd.to_datetime(news_sentiment['date']) + pd.Timedelta(days = 1)
```

```python
stock_prices['date'] = pd.to_datetime(stock_prices['date'])
```

```python
merged = pd.merge(stock_prices, news_sentiment, on = 'date', how = 'inner')
merged = merged[['stock', 'date', 'close', 'avg_sentiment']]
merged
```

```python
merged.groupby(['stock', 'date']).agg({
    'close':'mean',
    'avg_sentiment': 'mean'
}).reset_index()
```

```python
correlation = merged['close'].corr(merged['avg_sentiment'])
```
